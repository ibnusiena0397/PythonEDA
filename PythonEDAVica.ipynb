#Kelas Python EDA

##Import Library

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from google.colab import files
```

##Connect Datasource

```python
url = 'https://raw.githubusercontent.com/KennethParulian/PythonEDAEnigmacamp/main/MOCK_DATA%20(4).csv'
df = pd.read_csv(url)
```

##Data Validation

```python
df.head(20)
```

```python
df.info ()
```

```python
df.duplicated(subset=None, keep='first')
```

##Add Column

```python
df ['CreditRemaining '] = np.random.randint(1000, 15000, df.shape[0])
df.head()
```

```python
aa_milne_arr = ["Kansas", "LosAngles", "NewYork", "NewMexico"]
df ["City"] = np.random.choice(aa_milne_arr, 1000, df.shape[0])
df.head()
```

##Drop Duplicate Data

```python
df.drop("email", axis=1, inplace=True)
```

```python
df.head (10)
```

```python
print(df.isnull().sum())
```

```python
df.dropna(inplace=True)
df.count()
```

```python
df["CreditCardLimit"] =df["CreditCardLimit"].replace("[\$\,\.]","",regex=True).astype(int)
df.head()
```

##Rename Column

```python
df = df.rename(columns={"first_name": "FirstName", "last_name": "LastName", "gender":"Gender"})
```

```python
df.head(10)
```

```python
CustomerReport = df.groupby(["FirstName"])["CreditCardLimit"].sum().nlargest(8)

```

```python
plt.barh(CustomerReport.index, CustomerReport)
plt.grid(linestyle='--', linewidth=2, axis='y',alpha = 0.3)
plt.ylabel("LimitCreditCard")
plt.xlabel("CustomerName")
plt.title("TargetPromotion")
plt.show()
```

```python
Location = df.groupby(["City"])["id"].count()
```

```python
plt.bar(Location.index, Location)
plt.grid(linestyle='--', linewidth=2, axis='y',alpha = 0.3)
plt.ylabel("Person")
plt.xlabel("City")
plt.title("CustomerLocation")
plt.show()
```
